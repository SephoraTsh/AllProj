# Region of Wells

```python
central_longitude = selected_metadata['Longitude'].median()
average_cost = selected_metadata['TotalWellCost_USDMM'].mean()

# Classificando os poços como 'East' ou 'West' com base no ponto central da longitude.
selected_metadata['Region'] = selected_metadata['Longitude'].apply(lambda x: 'East' if x > central_longitude else 'West')

# Filtrando os poços que estão abaixo da média de custo.
below_average_cost = selected_metadata[selected_metadata['TotalWellCost_USDMM'] < average_cost]

# Contando o número de poços abaixo da média de custo por região.
wells_below_average_by_region = below_average_cost['Region'].value_counts()

# Criando o gráfico de barras.
wells_below_average_by_region.plot(kind='bar', color=['cyan', 'red'], figsize=(10, 6))

# Adicionando títulos e rótulos.
plt.title('Number of Wells Below Average Cost by Region', fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Number of Wells Below Average Cost')
plt.xticks(rotation=0)  # Os rótulos do eixo X serão 'East' e 'West'
plt.show()
```
